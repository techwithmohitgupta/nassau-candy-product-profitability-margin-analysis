## 1. Project Background

Nassau Candy Distributor operates as a large-scale distributor of confectionery and specialty food products across multiple regions. The company manages a diverse portfolio of products across different divisions such as Chocolate, Sugar, and Other categories.

While the company tracks sales performance effectively, relying solely on revenue can be misleading. High sales volume does not always translate into strong profitability. Some products may generate significant revenue but contribute very little to overall profit due to high production or operational costs.

This creates a critical need to evaluate business performance from a profitability perspective rather than just revenue.

## 2. Problem Statement

The organization currently lacks clear visibility into:

1. Which products generate the highest profit or gross margin

2. Whether high-sales products are actually profitable

3. How profitability varies across product divisions

4. Which products represent margin risk due to high cost or low efficiency

Without this insight, business decisions related to pricing, promotions, sourcing, and product portfolio optimization remain reactive and not data-driven.

## 3. Project Objective

The primary objective of this project is to:

Analyze product-level and division-level profitability

Evaluate financial performance using key metrics such as:
1. Gross Margin (%)
2. Profit per Unit
3. Revenue Contribution
4. Profit Contribution
5. Margin Volatility
   
Identify high-performing products and hidden profit drivers

Detect low-margin, cost-heavy, and inefficient products

Provide data-driven recommendations to improve profitability and optimize business strategy

## 4. Business Questions

This analysis aims to answer the following key business questions:

### Product-Level Questions

1. Which products generate the highest total profit?
2. Which products have the highest margin efficiency?
3. Are there products with high sales but low profitability?
4. Which products are underperforming financially?

### Division-Level Questions

1. Which divisions are most profitable?
2. Which divisions have strong revenue but weak margins?
3. Are there structural inefficiencies at the division level?

### Cost & Pricing Questions
1. Which products have disproportionately high cost?
2. Where are pricing inefficiencies present?
3. Which products require cost optimization or repricing?

### Risk & Dependency Questions
1. Is the business dependent on a small number of products?
2. What percentage of products contribute to 80% of revenue and profit?
3. Which products represent margin risk?

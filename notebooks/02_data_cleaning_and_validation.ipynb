## Import Libraries

```python
# Import required libraries

import pandas as pd
import numpy as np

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", "{:,.2f}".format)

print("Libraries imported successfully.")
```

## 🔷 1. Load Raw Dataset

We will load the raw dataset from the `data/raw` folder.

```python
# Load raw dataset

file_path = "../data/raw/Nassau Candy Distributor.csv"

df = pd.read_csv(file_path)

print("Raw dataset loaded successfully.")
print("Shape:", df.shape)
```

## 🔷 2. Initial Data Validation

Before cleaning, we validate the dataset structure, data types, and basic integrity.

```python
# Check dataset structure

df.info()
```

```python
# Check missing values

df.isnull().sum()
```

```python
# Check duplicate rows

df.duplicated().sum()
```

## 🔷 3. Text Cleaning and Standardization

In this step, we clean text-based columns by:

- Removing leading and trailing spaces
- Standardizing formatting
- Ensuring consistency across categories

```python
# Strip spaces from text columns

text_columns = [
    "Product Name",
    "Division",
    "City",
    "State/Province",
    "Region"
]

for col in text_columns:
    df[col] = df[col].str.strip()

print("Text columns cleaned successfully.")
```

```python
# Check unique values after cleaning

df["Division"].unique()
```

```python
df["City"].unique()
```

## 🔷 4. Product Name Standardization

In this step, we ensure that product names are fully standardized and consistent for accurate grouping and analysis.

This includes:

- Checking unique product count
- Verifying naming consistency
- Ensuring alignment with product categories

```python
# Count unique products

df["Product Name"].nunique()
```

```python
# Display all unique product names

df["Product Name"].unique()
```

## 🔷 5. Date Conversion

In this step, we convert date columns into proper datetime format for further analysis.

```python
# Convert date columns to datetime using day-first format

df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)

print("Date columns converted successfully.")
```

```python
# Verify the top rows of order date and ship date format 

df[["Order Date", "Ship Date"]].head()
```

```python
df[["Order Date", "Ship Date"]].dtypes
```

## 🔷 6. Financial Data Validation

In this step, we verify whether the financial data is consistent.

We check:
- Whether Gross Profit = Sales - Cost
- Ensure data integrity before analysis

```python
# Check financial consistency

df["Profit Check"] = df["Sales"] - df["Cost"] - df["Gross Profit"]

df["Profit Check"].describe()
```

```python
# Count mismatched rows

(df["Profit Check"] != 0).sum()
```

```python
df[df["Profit Check"] != 0].head()
```

```python
df["Profit Check"].abs().max()
```

```python
# Correct mismatch check using tolerance

(df["Profit Check"].abs() > 1e-6).sum()
```

```python
df[df["Profit Check"].abs() > 1e-6].head()
```

### Financial Validation:

- Profit calculation is consistent across dataset
- Minor floating-point differences observed due to computation precision
- No real financial mismatches found

## 🔷 7. Final Clean Dataset Preparation

In this step, we finalize the cleaned dataset by:

- Removing temporary validation columns
- Verifying final dataset structure
- Exporting the cleaned dataset for further analysis

```python
# Remove temporary column

df.drop(columns=["Profit Check"], inplace=True)

print("Temporary columns removed.")
```

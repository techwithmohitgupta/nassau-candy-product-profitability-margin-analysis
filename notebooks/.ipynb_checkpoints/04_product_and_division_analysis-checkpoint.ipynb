## 🔷 2. Product-Level Aggregation

In this step, we aggregate data at the product level to evaluate overall performance.

We calculate:

- Total Sales
- Total Profit
- Total Units
- Average Margin
- Average Profit per Unit

```python
# Product-level aggregation

product_df = df.groupby("Product Name").agg({
    "Sales": "sum",
    "Cost": "sum",   # 👈 ADD THIS
    "Gross Profit": "sum",
    "Units": "sum",
    "Gross Margin %": "mean",
    "Profit per Unit": "mean"
}).reset_index()

product_df.head()
```

### Add Contribution Metrics (Product-Level)

```python
# Recalculate contributions at product level

total_sales = product_df["Sales"].sum()
total_profit = product_df["Gross Profit"].sum()

product_df["Revenue Contribution %"] = product_df["Sales"] / total_sales
product_df["Profit Contribution %"] = product_df["Gross Profit"] / total_profit
```

## 🔷 4. Segment-Level Performance Analysis

In this step, we analyze product performance based on predefined segments.

Segments:

- ⭐ Stars → High Profit + High Margin
- 💎 Hidden Gems → High Profit + Low Margin
- ⚠️ Volume Traps → High Sales but Low Profit
- ❌ Underperformers → Low Sales + Low Profit

This analysis helps identify:

- Where the company is making money
- Where inefficiencies exist
- Which products require action

### Segment Distribution

```python
df["Product Segment"].value_counts()
```

### Segment-Level Aggregation

```python
segment_df = df.groupby("Product Segment").agg({
    "Sales": "sum",
    "Gross Profit": "sum",
    "Units": "sum",
    "Gross Margin %": "mean",
    "Profit per Unit": "mean"
}).reset_index()

segment_df
```

### Add Contribution Metrics

```python
total_sales = segment_df["Sales"].sum()
total_profit = segment_df["Gross Profit"].sum()

segment_df["Revenue Contribution %"] = segment_df["Sales"] / total_sales
segment_df["Profit Contribution %"] = segment_df["Gross Profit"] / total_profit

segment_df
```

### 🔷 Segment Performance Insights

⭐ Stars:
These products contribute the highest share of both revenue and profit, making them the backbone of the business. These should be prioritized for scaling and promotion.

💎 Hidden Gems:
These products generate strong profit but operate at lower margins. There may be pricing inefficiencies, indicating an opportunity to improve margins.

⚠️ Volume Traps:
These products generate high sales but low profit, indicating cost-heavy or inefficient pricing strategies. These require urgent optimization.

❌ Underperformers:
These products contribute minimal revenue and profit. They may need discontinuation or repositioning.

### 🔷 Final Segment Interpretation

⭐ Stars dominate both revenue and profit, indicating strong product-market fit and efficient pricing.

💎 Hidden Gems show high profitability but relatively lower margin efficiency, suggesting potential for margin optimization.

⚠️ Volume Traps generate significant sales but low profit, indicating cost inefficiencies or poor pricing strategies. These products require urgent optimization.

❌ Underperformers contribute the least to both revenue and profit, making them the weakest segment. These products should be evaluated for discontinuation or repositioning.

## 🔷 5. Division-Level Performance Analysis

In this step, we analyze profitability performance across product divisions.

This helps identify:

- Which division generates the highest revenue
- Which division contributes the most profit
- Which division has stronger margin efficiency
- Whether any division shows revenue-profit imbalance

### Division-Level Aggregation

```python
division_df = df.groupby("Division").agg({
    "Sales": "sum",
    "Gross Profit": "sum",
    "Cost": "sum",
    "Units": "sum",
    "Gross Margin %": "mean",
    "Profit per Unit": "mean"
}).reset_index()

division_df
```

### Add Contribution Metrics

```python
total_sales = division_df["Sales"].sum()
total_profit = division_df["Gross Profit"].sum()

division_df["Revenue Contribution %"] = division_df["Sales"] / total_sales
division_df["Profit Contribution %"] = division_df["Gross Profit"] / total_profit

division_df
```

### Sort by Profit

```python
division_df.sort_values(by="Gross Profit", ascending=False)
```

### 🔷 Division-Level Insights

🍫 Chocolate Division:
This division dominates both revenue (~93%) and profit (~95%), making it the primary driver of overall business performance. It shows strong scalability and should be the core focus for growth and expansion.

📦 Other Division:
Although this division contributes a smaller share of revenue (~7%), it has the highest profit per unit, indicating strong margin efficiency. This suggests premium or high-margin products that could be scaled strategically.

🍬 Sugar Division:
This division contributes negligible revenue and profit, making it the weakest performer. It may require strategic review, including pricing optimization, cost control, or potential discontinuation of low-performing products.

⚠️ Key Business Insight:
The business is highly dependent on the Chocolate division, indicating a concentration risk. Diversification across other divisions is recommended to reduce dependency and improve overall stability.

## 🔷 7. Pareto Analysis (80/20 Rule)

In this step, we identify the concentration of revenue and profit across products.

This helps determine:

- Which products drive the majority of revenue
- Which products drive the majority of profit
- Whether the business is dependent on a few key products

### Sort by Revenue

```python
pareto_sales = product_df.sort_values(by="Sales", ascending=False).copy()

pareto_sales["Cumulative Sales %"] = pareto_sales["Sales"].cumsum() / pareto_sales["Sales"].sum()

pareto_sales
```

### Identify 80% Revenue Products

```python
top_80_sales = pareto_sales[pareto_sales["Cumulative Sales %"] <= 0.80]

top_80_sales
```

### Sort by Profit

```python
pareto_profit = product_df.sort_values(by="Gross Profit", ascending=False).copy()

pareto_profit["Cumulative Profit %"] = pareto_profit["Gross Profit"].cumsum() / pareto_profit["Gross Profit"].sum()

pareto_profit
```

### Identify 80% Profit Products

```python
top_80_profit = pareto_profit[pareto_profit["Cumulative Profit %"] <= 0.80]

top_80_profit
```

### 🔷 Pareto Insights (80/20 Rule)

Revenue Concentration:
A small number of products contribute to the majority of total revenue, indicating a concentration of business performance around key products.

Profit Concentration:
Similarly, a limited set of products generates the majority of total profit, highlighting the most valuable products for the company.

⚠️ Business Risk:
The company is heavily dependent on a few products for both revenue and profit, creating a dependency risk.

💡 Recommendations:
- Strengthen top-performing products (Stars)
- Reduce dependency by improving performance of other products
- Optimize underperforming products to balance portfolio

### 🔷 Pareto Analysis Insights (80/20 Rule)

Revenue Concentration:
Approximately 80% of total revenue is generated by just **2 products**, indicating a strong concentration of sales within a very small subset of the product portfolio.

Profit Concentration:
Even more critically, **80% of total profit is driven by only 1 product**, highlighting extreme dependency on a single high-performing product.

⚠️ Business Risk:
This indicates a **high concentration risk**, where the business is heavily dependent on a limited number of products—especially one key product—for profitability.

Any disruption in demand, pricing, or supply of this product could significantly impact overall business performance.

💡 Strategic Recommendations:
- Strengthen and protect top-performing products (especially the top profit driver)
- Reduce dependency by improving performance of mid-tier and underperforming products
- Focus on converting "Hidden Gems" into "Stars"
- Re-evaluate low-margin products (Volume Traps) for cost optimization or discontinuation

## 🔷 8. Export Analysis Datasets

In this step, we export all analysis-ready datasets generated in Notebook 04.

These files will be saved inside the `data/processed` folder and will be used for dashboard development, reporting, and final business storytelling.

Exported datasets include:

- Final feature-engineered dataset
- Product-level analysis
- Division-level analysis
- Top sales products
- Top profit products
- Bottom products
- Segment-level performance
- Cost vs profit diagnostics
- Pareto sales analysis
- Pareto profit analysis

```python
# =========================================
# EXPORT ANALYSIS DATASETS
# =========================================

import os

# Create processed folder if it does not exist
os.makedirs("../data/processed", exist_ok=True)

# -----------------------------------------
# 1. Main datasets
# -----------------------------------------

df.to_csv("../data/processed/final_feature_engineered_dataset.csv", index=False)
product_df.to_csv("../data/processed/product_level_analysis.csv", index=False)
division_df.to_csv("../data/processed/division_level_analysis.csv", index=False)

# -----------------------------------------
# 2. Product ranking datasets
# -----------------------------------------

top_sales = product_df.sort_values(by="Sales", ascending=False)
top_profit = product_df.sort_values(by="Gross Profit", ascending=False)
bottom_products = product_df.sort_values(by="Gross Profit", ascending=True)

top_sales.to_csv("../data/processed/top_sales_products.csv", index=False)
top_profit.to_csv("../data/processed/top_profit_products.csv", index=False)
bottom_products.to_csv("../data/processed/bottom_products.csv", index=False)

# -----------------------------------------
# 3. Segment-level performance dataset
# -----------------------------------------

segment_summary = df.groupby("Product Segment").agg({
    "Sales": "sum",
    "Cost": "sum",
    "Gross Profit": "sum",
    "Units": "sum",
    "Gross Margin %": "mean",
    "Profit per Unit": "mean"
}).reset_index()

segment_summary["Revenue Contribution %"] = segment_summary["Sales"] / segment_summary["Sales"].sum()
segment_summary["Profit Contribution %"] = segment_summary["Gross Profit"] / segment_summary["Gross Profit"].sum()

segment_summary.to_csv("../data/processed/segment_performance.csv", index=False)

# -----------------------------------------
# 4. Cost vs profit diagnostics dataset
# -----------------------------------------

cost_profit_df = product_df.copy()

cost_profit_df["Cost Ratio"] = cost_profit_df["Cost"] / cost_profit_df["Sales"]

cost_profit_df = cost_profit_df.sort_values(by="Cost Ratio", ascending=False)

cost_profit_df.to_csv("../data/processed/cost_vs_profit_analysis.csv", index=False)

# -----------------------------------------
# 5. Pareto analysis datasets
# -----------------------------------------

pareto_sales.to_csv("../data/processed/pareto_sales_analysis.csv", index=False)
pareto_profit.to_csv("../data/processed/pareto_profit_analysis.csv", index=False)

# -----------------------------------------
# Final confirmation
# -----------------------------------------

print("✅ All Notebook 04 analysis datasets exported successfully to data/processed/")
```

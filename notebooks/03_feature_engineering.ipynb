## 🔷 Objective

In this notebook, we create key financial and analytical metrics required for profitability analysis.

This includes:

- Gross Margin (%)
- Profit per Unit
- Revenue Contribution
- Profit Contribution

These features will help evaluate both financial performance and efficiency.

### Import Libraries

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

### Load Clean Dataset

```python
file_path = "../data/processed/cleaned_nassau_candy.csv"

df = pd.read_csv(file_path)

print("Clean dataset loaded.")
df.head()
```

### Convert Date Again (Important)

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

## 🔷 1. Gross Margin (%)

Gross Margin measures profitability efficiency.

Formula:
Gross Margin = Gross Profit / Sales

```python
df["Gross Margin %"] = df["Gross Profit"] / df["Sales"]
```

```python
df[["Sales", "Gross Profit", "Gross Margin %"]].head()
```

## 🔷 2. Profit per Unit

Profit per Unit measures how much profit is generated per unit sold.

Formula:
Profit per Unit = Gross Profit / Units

```python
df["Profit per Unit"] = df["Gross Profit"] / df["Units"]
```

```python
df[["Units", "Gross Profit", "Profit per Unit"]].head()
```

## 🔷 3. Revenue Contribution (%)

Revenue Contribution measures how much each product contributes to total sales.

Formula:
Revenue Contribution = Product Sales / Total Sales

```python
# Calculate total sales

total_sales = df["Sales"].sum()

# Create revenue contribution column

df["Revenue Contribution %"] = df["Sales"] / total_sales
```

### Verify

```python
df[["Sales", "Revenue Contribution %"]].head()
```

## 🔷 4. Profit Contribution (%)

Profit Contribution measures how much each product contributes to total profit.

Formula:
Profit Contribution = Product Profit / Total Profit

```python
# Calculate total profit

total_profit = df["Gross Profit"].sum()

# Create profit contribution column

df["Profit Contribution %"] = df["Gross Profit"] / total_profit
```

```python
df[["Gross Profit", "Profit Contribution %"]].head()
```

## 🔷 5. Margin Volatility

Margin Volatility measures how much profit margin fluctuates over time.

Approach:
- Calculate margin at time level (monthly)
- Measure variability using standard deviation

### Create Month Column

```python
# Extract month from Order Date

df["Order Month"] = df["Order Date"].dt.to_period("M")
```

### Monthly Margin Calculation

```python
# Calculate monthly sales and profit

monthly_df = df.groupby("Order Month").agg({
    "Sales": "sum",
    "Gross Profit": "sum"
}).reset_index()

# Calculate monthly margin

monthly_df["Monthly Margin"] = monthly_df["Gross Profit"] / monthly_df["Sales"]

monthly_df.head()
```

### Margin Volatility (Final KPI)

```python
# Calculate margin volatility

margin_volatility = monthly_df["Monthly Margin"].std()

print("Margin Volatility:", margin_volatility)
```

### Observation

The calculated Margin Volatility is very low, indicating that the company's gross margin remains stable across months. This suggests consistent profitability efficiency over time and lower margin instability risk at the overall business level.

## 🔷 7. Segmentation Thresholds

We define thresholds using median values:

- Profit Threshold → Median Profit per Unit
- Margin Threshold → Median Gross Margin %

```python
# Define thresholds

profit_threshold = df["Profit per Unit"].median()
margin_threshold = df["Gross Margin %"].median()

print("Profit Threshold:", profit_threshold)
print("Margin Threshold:", margin_threshold)
```

## 🔷 8. Product Segmentation

Products are categorized based on:

- Profit per Unit (High / Low)
- Gross Margin % (High / Low)

Segmentation Categories:

- ⭐ Stars → High Profit + High Margin  
- ⚠️ Volume Traps → High Profit + Low Margin  
- 💎 Hidden Gems → Low Profit + High Margin  
- ❌ Underperformers → Low Profit + Low Margin  

```python
# Create segmentation logic

def classify_product(row):
    if row["Profit per Unit"] >= profit_threshold and row["Gross Margin %"] >= margin_threshold:
        return "Stars"
    
    elif row["Profit per Unit"] >= profit_threshold and row["Gross Margin %"] < margin_threshold:
        return "Volume Traps"
    
    elif row["Profit per Unit"] < profit_threshold and row["Gross Margin %"] >= margin_threshold:
        return "Hidden Gems"
    
    else:
        return "Underperformers"

# Apply classification

df["Product Segment"] = df.apply(classify_product, axis=1)

print("Segmentation applied successfully.")
```

### Verify Segments

```python
df["Product Segment"].value_counts()
```

### 🔷 Segmentation Insights

- A significant portion of products fall under the "Stars" category, indicating strong profitability and margin performance.
- However, a large number of "Underperformers" suggests inefficiencies in certain product lines that require strategic review.
- "Hidden Gems" represent high-margin products with lower sales volume, indicating potential for growth through better marketing or distribution.
- "Volume Traps" highlight products generating profit but operating at lower margins, suggesting potential cost or pricing issues.
